## Data Exploration

```python
data0.isna().any()
```



The data has no missing values but some missing values have been made 0. This can be concluded as many columns has minimum value as 0.
No living person can have zero insulin or blood pressure at 0, and neither can their BMI. So we will have to replace these values with some more appropriate ones.



```python
# We are gonna replace the 0 with random value around the mean with standard Deviation of 1.2 (+/-)

for col in ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]:
    mean = data0[data0[col] != 0][col].mean()
    std = data0[data0[col] != 0][col].std()

    values = data0[col].values

    np.random.seed(23)

    for i, val in enumerate(values):
        if val == 0:
            values[i] = mean + std * (np.random.rand() * 2 - 1) * 1.2
    data0[col] = pd.Series(values).astype(data0[col].dtype)
```

## Spliting the data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python

X = data1.drop("Outcome", axis=1)
y = data1["Outcome"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)

ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Searching for baseline model

```python
from sklearn.model_selection import cross_val_score, GridSearchCV

from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, roc_auc_score
```

```python
def cv_and_evaluate(model, X_train, y_train, X_test, y_test, name=None):

    cv_score = cross_val_score(model, X_train, y_train, cv=4, n_jobs=-1)
    
    print(f"{name} CV Results \t Mean: {np.mean(cv_score)} \t Std: {np.std(cv_score)}")
    print(cv_score)

    model.fit(X_train, y_train)

    y_test_pred = model.predict(X_test)
    
#     try:
#         y_test_score = model.predict_proba(X_test)[:, 1]
#     except:
#         y_test_score = model.decision_function(X_test)

    #classifier_evaluation(y_test, y_test_pred, y_test_score)
    
    accuracy = accuracy_score(y_test, y_test_pred)
    precision = precision_score(y_test, y_test_pred)
    recall  = recall_score(y_test, y_test_pred)
    f1 = f1_score(y_test, y_test_pred)
    
    print("Accuracy = ", np.round(accuracy,2))
    print("Precision = ", np.round(precision,2))
    print("Recall = ", np.round(recall,2))
    print("F1 = ", np.round(f1,2))
```

```python
# Adaboost

base = DecisionTreeClassifier(random_state=23, max_depth=3)
adb_clf = AdaBoostClassifier(base, random_state=23, n_estimators=50)

cv_and_evaluate(adb_clf, X_train, y_train, X_test, y_test, name="AdaBoost")
```

```python
# Logistic regression

lr = LogisticRegression()

cv_and_evaluate(lr, X_train_scaled, y_train, X_test_scaled, y_test, name="LogisticRegression")
```

```python
# RandomForest

rf = RandomForestClassifier(max_depth=8, random_state=23)

cv_and_evaluate(rf, X_train, y_train, X_test, y_test, name="Random Forest")
```

```python
#SVC

svc = SVC(probability=True, random_state=23)

cv_and_evaluate(svc, X_train_scaled, y_train, X_test_scaled, y_test, name="SVC")
```



The results are very similar to each other.
We are choosing SVC as the model with which to proceed. SVM allow good tuning, and having little data they perform very well, both in time factor and metrics



## Hyperparameter Tuning

We will use different kernel to see if there is any difference.

```python
model = SVC(kernel="poly", degree=3, random_state=23)

cv_and_evaluate(model, X_train_scaled, y_train, X_test_scaled, y_test, name="SVC Poly")
```

```python
model = SVC(kernel="linear", random_state=23)

cv_and_evaluate(model, X_train_scaled, y_train, X_test_scaled, y_test, name="SVC Linear")
```



Looking at the results, we are keeping the default kernel. 
In fact, having to diagnose a disease, it is more important for our model to score a higher recall score, even at the cost of giving up some precision.


```python


model = SVC(probability=True, random_state=23)

params_grid = {
    "gamma": [0.1, 1, 10, 100],
    "C": [0.1, 1, 10, 100, 1000]
}

svc_gridsearch = GridSearchCV(model, params_grid, n_jobs=-1)

svc_gridsearch.fit(X_train_scaled, y_train)

```

```python

print("Best params: \t", svc_gridsearch.best_params_)
print("Best Score: \t", svc_gridsearch.best_score_)
```

```python
model = svc_gridsearch.best_estimator_

cv_and_evaluate(model, X_train_scaled, y_train, X_test_scaled, y_test, name="SVC tuned")
```


There is no much difference even after applying hyperparameter.


## Feature Selection

We will selecting only important features for better results.

```python


from sklearn.inspection import permutation_importance

importances = pd.Series(permutation_importance(model, X_test_scaled, y_test, random_state=23).importances_mean, 
                        index=X_test.columns)

importances.sort_values(ascending=False, inplace=True)

fig, ax = plt.subplots(figsize=(12, 6))

sns.barplot(x=importances.values, y=importances.index, ax=ax, linewidth=2, edgecolor="black")

ax.set_xlabel("Permutation Importance")
ax.set_title("Feature Importances - SVC")

sns.despine(ax=ax, left=True, bottom=False)
ax.grid(axis="y", linewidth=0.2, color="black")
#set_style(ax, axis_grid="x", bottom=True)

plt.show()

```

In this notebook we carried out in-depth data exploration, making the best use of data visualization.
We also cleaned the data, replacing null values with more probable ones, without losing any training instances.
Later, after finding a promising model, we optimized it. To achieve even better performance, we finally oversampled the minority class, and selected the best features.

The results are not the best possible, and this is due to the small amount of data available. By having more training instances available most likely our model could have had much better results.
